Principal Component Analysis

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
```

```python
breast.data.shape
```

```python
breast_labels = breast.target
```

```python
breast_labels.shape
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
final_breast_data = np.concatenate([breast_data,labels],axis=1)
```

```python
final_breast_data.shape
```

```python
import pandas as pd
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
features = breast.feature_names
```

```python
features
```

```python
features_labels = np.append(features,'label')
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
```

```python
#breast_dataset.head()// use to display the few raw from the begining
#tail use to display the last 5 raw from the dataset
breast_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) 
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['PCA component 1', 'PCA component 2'])
```

```python
principal_breast_Df.tail()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
from keras.datasets import cifar10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print('Traning data shape:', x_train.shape)
print('Testing data shape:', x_test.shape)
```

```python
y_train.shape,y_test.shape
```

```python
# Find the unique numbers from the train labels
classes = np.unique(y_train)
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

```python
label_dict = {
 0: 'airplane',
 1: 'automobile',
 2: 'bird',
 3: 'cat',
 4: 'deer',
 5: 'dog',
 6: 'frog',
 7: 'horse',
 8: 'ship',
 9: 'truck',
}
```

```python
label_dict
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
np.min(x_train),np.max(x_train)
```

```python
x_train.shape
```

```python
x_train_flat = x_train.reshape(-1,3072)
x_test_flat = x_test.reshape(-1,3072)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.9)
```

```python
pca.fit(x_train_flat)
```

```python
pca.n_components_
```

```python
train_img_pca = pca.transform(x_train_flat)
test_img_pca = pca.transform(x_test_flat)
```

```python
import tensorflow
from tensorflow.keras.optimizers import RMSprop
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils

```

```python
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
batch_size = 128
num_classes = 10
epochs = 20
```

```python
model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(99,)))
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(train_img_pca, y_train,batch_size=batch_size,epochs=epochs,verbose=1,
                    validation_data=(test_img_pca, y_test))
```

```python
model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(3072,)))
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(x_train_flat, y_train,batch_size=batch_size,epochs=epochs,verbose=1,
                    validation_data=(x_test_flat, y_test))

```
